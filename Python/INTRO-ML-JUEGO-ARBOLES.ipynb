## Idea clave (sin tecnopalabras)

Piensa en un **mapa de preguntas**:
- Tú respondes sí/no.
- Cada respuesta te manda por un camino.
- Al final el programa se moja y adivina.

Ese "ir tomando decisiones por pasos" ya se parece bastante a cómo funcionan muchos modelos.

```python
def recorrer(arbol, respuestas):
    """Recorre el mapa con una lista de respuestas: ["si", "no", ...]"""
    actual = arbol
    i = 0

    while "pregunta" in actual:
        if i >= len(respuestas):
            return "Faltan respuestas para continuar"

        r = respuestas[i].strip().lower()
        if r not in ("si", "no"):
            return f"Respuesta no válida en posición {i}: {respuestas[i]}"

        actual = actual[r]
        i += 1

    return actual["adivinanza"]

print("Ruta ['si'] ->", recorrer(arbol, ["si"]))
print("Ruta ['no', 'si'] ->", recorrer(arbol, ["no", "si"]))
print("Ruta ['no', 'no'] ->", recorrer(arbol, ["no", "no"]))
```

## ¿Dónde está el aprendizaje aquí?

Cuando el programa se equivoca:
1. Le dices cuál era el objeto de verdad.
2. Le das una pregunta que lo distinga del que dijo mal.
3. El mapa se actualiza.

Traducción simple: **falla → corriges → mejora**.

```python
def aprender_simple(nodo_hoja, nuevo_objeto, nueva_pregunta, respuesta_nuevo):
    """
    Actualiza una hoja con una nueva pregunta para separar
    el objeto fallido del nuevo objeto correcto.
    """
    anterior = nodo_hoja["adivinanza"]
    viejo = {"adivinanza": anterior}
    nuevo = {"adivinanza": nuevo_objeto}

    nodo_hoja.clear()
    nodo_hoja["pregunta"] = nueva_pregunta

    if respuesta_nuevo == "si":
        nodo_hoja["si"] = nuevo
        nodo_hoja["no"] = viejo
    else:
        nodo_hoja["si"] = viejo
        nodo_hoja["no"] = nuevo

# Ejemplo guiado: en la rama 'si' antes estaba 'perro'
hoja_animal = arbol["si"]
aprender_simple(
    hoja_animal,
    nuevo_objeto="gato",
    nueva_pregunta="¿Maúlla?",
    respuesta_nuevo="si"
)

arbol
```

### Paso 1 · qué haces aquí

- **Por qué:** la gente responde `sí`, `si`, `s`, `no`...
- **Cómo:** conviertes todo a `si` o `no`.
- **Qué ganas:** el juego no falla por variaciones de escritura.

✅ Este paso hace la entrada robusta.

```python
# Paso 1: funciones base para trabajar con respuestas sí/no

def normalizar_si_no(texto):
    texto = texto.strip().lower()
    if texto in {"s", "si", "sí", "y", "yes"}:
        return "si"
    if texto in {"n", "no"}:
        return "no"
    return ""


def pedir_si_no(pregunta):
    while True:
        r = input(f"{pregunta} (sí/no): ")
        rama = normalizar_si_no(r)
        if rama:
            return rama
        print("Ups, responde con sí o no ✋")
```

### Paso 2 · memoria del juego

- **Por qué:** sin guardar, se pierde lo aprendido.
- **Cómo:** guardas/cargas el árbol en JSON.
- **Qué pasa:** al cerrar se escribe en disco; al abrir se recupera.

✅ JSON = cuaderno de memoria del juego.

```python
# Paso 2: carga y guardado del conocimiento (persistencia)
import json
from pathlib import Path

RUTA_JUEGO = Path("arbol_juego_clase.json")
ARBOL_BASE = {
    "pregunta": "¿Es un animal?",
    "si": {"adivinanza": "perro"},
    "no": {
        "pregunta": "¿Es un vegetal?",
        "si": {"adivinanza": "lechuga"},
        "no": {"adivinanza": "piedra"},
    },
}


def cargar_arbol_clase():
    if not RUTA_JUEGO.exists():
        return ARBOL_BASE
    with RUTA_JUEGO.open("r", encoding="utf-8") as f:
        data = json.load(f)
    return data if isinstance(data, dict) else ARBOL_BASE


def guardar_arbol_clase(arbol_actual):
    with RUTA_JUEGO.open("w", encoding="utf-8") as f:
        json.dump(arbol_actual, f, ensure_ascii=False, indent=2)

print("Ruta de guardado:", RUTA_JUEGO.resolve())
```

### Paso 3 · dónde aprende

- **Por qué:** no basta con fallar; queremos mejorar.
- **Cómo:** pides objeto correcto + pregunta que diferencia.
- **Qué cambia:** una hoja pasa a ser una nueva bifurcación.

✅ El árbol crece con feedback humano.

```python
# Paso 3: aprendizaje cuando falla

def aprender_en_nodo_clase(nodo):
    fallida = nodo["adivinanza"]

    nuevo_objeto = input("Me rindo 😅 ¿Qué objeto era?: ").strip()
    while not nuevo_objeto:
        nuevo_objeto = input("Escribe un objeto válido: ").strip()

    nueva_pregunta = input(
        f"Dame una pregunta para diferenciar '{nuevo_objeto}' de '{fallida}': "
    ).strip()
    while not nueva_pregunta:
        nueva_pregunta = input("La pregunta no puede estar vacía. Prueba otra vez: ").strip()

    respuesta_nuevo = pedir_si_no(
        f"Para '{nuevo_objeto}', ¿la respuesta a esa pregunta es sí?"
    )

    viejo = {"adivinanza": fallida}
    nuevo = {"adivinanza": nuevo_objeto}

    nodo.clear()
    nodo["pregunta"] = nueva_pregunta
    if respuesta_nuevo == "si":
        nodo["si"] = nuevo
        nodo["no"] = viejo
    else:
        nodo["si"] = viejo
        nodo["no"] = nuevo
```

### Paso 4 · ronda completa

- **Por qué:** unir funciones sueltas en un juego real.
- **Cómo:** preguntar → adivinar → aprender si falla → repetir.
- **Qué pasa al final:** al salir, se guarda lo aprendido.

✅ Ciclo completo: entrada, decisión, feedback, memoria.

```python
# Paso 4: bucle principal del juego

def jugar_una_ronda(arbol_actual):
    actual = arbol_actual

    while "pregunta" in actual:
        rama = pedir_si_no(actual["pregunta"])
        actual = actual[rama]

    confirmar = pedir_si_no(f"¿Estás pensando en '{actual['adivinanza']}'?")
    if confirmar == "si":
        print("¡Te leí la mente! 😎")
        return

    aprender_en_nodo_clase(actual)
    print("Gracias, acabo de aprender algo nuevo ✨")


def ejecutar_juego_completo():
    print("\n🎮 Juego: Animal, vegetal o mineral")
    print("Piensa en algo y vamos a jugar...\n")

    arbol_actual = cargar_arbol_clase()

    while True:
        jugar_una_ronda(arbol_actual)
        otra = pedir_si_no("¿Jugamos otra ronda?")
        if otra == "no":
            guardar_arbol_clase(arbol_actual)
            print("Guardado listo. ¡Nos vemos! 👋")
            break
```

## 🔵 Bloque 3 (Puente a ML): de árbol a datos + predicción

Si ya conoces pandas, piensa así:
- Vas a convertir el árbol en una **tabla de ejemplos**.
- Cada fila será un caso (un camino de respuestas).
- Las columnas serán preguntas respondidas con 0/1.
- La última columna será el objeto final (la etiqueta).

👉 Traducción importante:
- **Tabla de ejemplos** = un `DataFrame` con filas de casos reales.
- **Clasificador** = una función/modelo que intenta adivinar la etiqueta final.

### Noción clave: ¿qué es una etiqueta?

Una **etiqueta** es la respuesta final que queremos predecir.

En este notebook:
- Las preguntas (sí/no) son los datos de entrada.
- La **etiqueta** es el objeto final (`perro`, `lechuga`, `piedra`, etc.).

Piensa en ello así:
- Entrada = pistas.
- Etiqueta = solución/categoría final.

Cuando veas `df["etiqueta"]`, esa columna guarda la respuesta correcta de cada fila.

### Noción clave: input y output

En cualquier problema de datos, piensa en este esquema:

- **Input (entrada):** la información que le das al sistema.
- **Output (salida):** lo que el sistema devuelve.

En este juego:
- Input = respuestas a preguntas (`sí/no`), que en la tabla se ven como columnas 0/1.
- Output = la predicción final (la **etiqueta**, como `perro` o `lechuga`).

Regla rápida para recordarlo:
**Input entra, output sale.**

### ¿De dónde sale y dónde entra todo esto?

Pregunta súper importante 👇

1. **¿De dónde salen los datos?**
   - Salen del árbol (`arbol`) que construiste antes.
   - La función `extraer_ejemplos(...)` recorre ese árbol y crea casos.

2. **¿Dónde entra el input?**
   - En modo juego: entra cuando tú respondes sí/no.
   - En modo tabla: entra en las columnas de preguntas (0/1).

3. **¿De dónde sale el output?**
   - Sale de la columna `etiqueta` (respuesta correcta).
   - Y en predicción, sale como la etiqueta que devuelve el modelo.

Piensa el flujo así:
**árbol → ejemplos → tabla (input + etiqueta) → predicción (output)**

```python
# 1) Pasar el árbol a ejemplos y verlo como tabla de pandas
from collections import Counter, defaultdict
import random
import pandas as pd


def extraer_ejemplos(nodo, camino=None):
    if camino is None:
        camino = []

    # Si llegamos a una adivinanza, tenemos un ejemplo completo
    if "adivinanza" in nodo:
        return [(camino, nodo["adivinanza"])]

    ejemplos_si = extraer_ejemplos(nodo["si"], camino + [(nodo["pregunta"], 1)])
    ejemplos_no = extraer_ejemplos(nodo["no"], camino + [(nodo["pregunta"], 0)])
    return ejemplos_si + ejemplos_no


ejemplos = extraer_ejemplos(arbol)

# "Vocabulario" de preguntas (serán columnas)
preguntas = sorted({pregunta for camino, _ in ejemplos for pregunta, _ in camino})


def vectorizar(camino, preguntas):
    mapa = {pregunta: valor for pregunta, valor in camino}
    return tuple(mapa.get(p, 0) for p in preguntas)


X = [vectorizar(camino, preguntas) for camino, _ in ejemplos]
y = [etiqueta for _, etiqueta in ejemplos]

# Construimos DataFrame (esto es la "tabla de ejemplos")
df = pd.DataFrame(X, columns=preguntas)
df["etiqueta"] = y

print("Tabla de ejemplos (DataFrame):")
display(df)

print("\nResumen:")
print(f"Filas (casos): {len(df)}")
print(f"Columnas de preguntas: {len(preguntas)}")
print("Columna objetivo: etiqueta")
```

### ¿Qué es un clasificador? (versión súper simple)

Un clasificador es una herramienta que mira datos de entrada y devuelve una categoría.

En este caso:
- **Entrada**: respuestas a preguntas (columnas 0/1).
- **Salida**: `etiqueta` (objeto final).

No vamos a usar un modelo complejo: haremos uno básico para entender la idea de predecir.

```python
# 2) Entrenar un clasificador simple por mayoría de patrones
# (si ve el mismo patrón de respuestas, predice la etiqueta más frecuente)

# Duplicamos ejemplos con pequeño ruido para tener un mini dataset de práctica
random.seed(42)
X_aug, y_aug = [], []
for features, etiqueta in zip(X, y):
    for _ in range(15):
        x = list(features)
        # Ruido muy pequeño: a veces invierte una feature
        if random.random() < 0.08 and len(x) > 0:
            j = random.randrange(len(x))
            x[j] = 1 - x[j]
        X_aug.append(tuple(x))
        y_aug.append(etiqueta)

# Split train/test sencillo
indices = list(range(len(X_aug)))
random.shuffle(indices)
corte = int(0.7 * len(indices))
idx_train = indices[:corte]
idx_test = indices[corte:]

X_train = [X_aug[i] for i in idx_train]
y_train = [y_aug[i] for i in idx_train]
X_test = [X_aug[i] for i in idx_test]
y_test = [y_aug[i] for i in idx_test]


def entrenar_modelo_mayoria(X_train, y_train):
    tabla = defaultdict(list)
    for x, etiqueta in zip(X_train, y_train):
        tabla[x].append(etiqueta)

    modelo = {x: Counter(labels).most_common(1)[0][0] for x, labels in tabla.items()}
    clase_global = Counter(y_train).most_common(1)[0][0]
    return modelo, clase_global


def predecir(modelo, clase_global, x):
    return modelo.get(x, clase_global)


modelo, clase_global = entrenar_modelo_mayoria(X_train, y_train)
preds = [predecir(modelo, clase_global, x) for x in X_test]
aciertos = sum(p == t for p, t in zip(preds, y_test))
acc = aciertos / len(y_test) if y_test else 0

print(f"Tamaño train: {len(X_train)} | test: {len(X_test)}")
print(f"Exactitud del mini-modelo: {acc:.2%}")
```
