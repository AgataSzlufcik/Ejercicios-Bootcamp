```python


"""
Ejercicio: comprobar si una matriz cuadrada es magica.

Instrucciones para alumnado:
1. Completa las funciones marcadas con TODO.
2. Usa return en vez de print para la logica.
3. Prueba con los casos de ejemplo al final.
"""

""" Necesito calcular todas las sumas de filas, columnas y diagonales, si son iguales la matriz será magica. MAtriz es una lista de listas"""

#tengo que acceder a la fila i y sumar todos sus componentes

def suma_fila(matriz, i):
    """Recibe una matriz y un índice de fila, y devuelve la suma de esa fila."""

    # Una fila es simplemente una lista dentro de la matriz
    # sum() recorre todos sus elementos y los suma. 

    return sum(matriz[i])
    pass


def suma_columna(matriz, j):
    """Recibe una matriz y un índice de columna, y devuelve la suma de esta columna ."""

    # Para recorrer una columna hay que ir fila a fila
    # y quedarse siempre con el elemento de la posición j. Empezamos por 0

    total = 0 
    for i in range(len(matriz)): 
        total += matriz[i][j] 
    return total
    pass


def suma_diagonal_principal(matriz):
    """Recibe una matriz y devuelve la suma de la diagonal principal."""

    # La diagonal principal es la diagonal que va desde la esquina superior izquierda a la esquina inferior derecha

    return sum((matriz[i][i]) for i in range(len(matriz)))

    pass


def suma_diagonal_secundaria(matriz):
    """Recibe una matriz y devuelve la suma de la diagonal secundaria."""
    
    # # La diagonal secundaria va de esquina superior-derecha a inferior-izquierda:
    # (0, len(matriz)-1), (1, len(matriz)-2), (2, n-3), ...
    # El índice de columna es siempre ((len(matriz) - 1 - i)).

    return sum((matriz[i][len(matriz) -1 -i]) for i in range(len(matriz)) )
    pass


def es_matriz_cuadrada(matriz):
    """Recibe una matriz y devuelve si es cuadrada o no (True/False)"""

    if not matriz:
        return False

    n = len(matriz)
    for fila in matriz:
        if len(fila) != n:
            return False
    return True


def es_matriz_magica(matriz):
    """Recibe una matriz y devuelve si es mágica o no."""

    # validar matriz cuadrada
    if not es_matriz_cuadrada(matriz):
        return False
    
    # calcular suma objetivo (ejemplo: primera fila). Esto es el “patrón” que todo tiene que cumplir. Coges la primera fila, la sumas, guardas ese número
    suma_objetivo = suma_fila(matriz,0)
    n = len(matriz)

    # comparar filas
    for i in range(n):
        if suma_fila(matriz, i) != suma_objetivo:
            return False
        
   # comparar columnas
    for j in range(n):
        if suma_columna(matriz, j) !=suma_objetivo:
            return False

    # comparar diagonales
    if suma_diagonal_principal(matriz) !=suma_objetivo:
        return False
    if suma_diagonal_secundaria(matriz) != suma_objetivo:
        return False     
    
    return True



def ejecutar_pruebas():
    matriz_magica_3x3 = [
        [8, 1, 6],
        [3, 5, 7],
        [4, 9, 2],
    ]

    matriz_no_magica = [
        [1, 2, 3],
        [4, 5, 6],
        [7, 8, 9],
    ]

    print("Caso magico esperado True:", es_matriz_magica(matriz_magica_3x3))
    print("Caso no magico esperado False:", es_matriz_magica(matriz_no_magica))


if __name__ == "__main__":
    ejecutar_pruebas()
```
