```python
df_csv = pd.read_csv("ventas_tienda_festival.csv")
df_csv["importe"] = df_csv["precio_eur"] * df_csv["unidades"]
filtrado = df_csv[(df_csv["importe"] >= 30)].copy() #hay que cerrar con [] en principio para crear la tabla, no valor Booleano True/False. esta es copia
#filtrado.to_html("mi_nuevo_archivo.html", index = False)

tabla_html = filtrado.to_html(index=False)

html = f"""
<!DOCTYPE html>
<html lang="es">
<head>
    <meta charset="UTF-8">
    <title>Ventas del festival</title>
</head>
<body>
    <h1>Ventas con importe mayor o igual a 30€</h1>
    {tabla_html}
</body>
</html>
"""

with open("mi_nuevo_archivo.html", "w") as file:
    file.write(html)

display(filtrado)
```
